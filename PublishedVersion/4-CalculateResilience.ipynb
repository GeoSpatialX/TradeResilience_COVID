```python
import pandas
```

## 1. Regional
### 1.1. Regions in global network

```python
regionalYoY = pandas.read_excel('./Tables/RegionalYoY.xlsx', sheet_name = 'RegionalYoY')
regionalYoY['yearmonth'] = (regionalYoY['year'] - 2020) * 12 + regionalYoY['month']

regionalRobust = regionalYoY.\
    groupby(['region']).\
    aggregate({
        'ex_vol': 'min',
        'im_vol': 'min'
    }).\
    eval(
        'ex_robust = 1 + ex_vol' '\n'
        'im_robust = 1 + im_vol'
    )

regionalResponseExWorst = regionalYoY.\
    iloc[regionalYoY.groupby('region')['ex_vol'].idxmin()].\
    set_index('region').\
    rename(columns = {
        'yearmonth': 'ex_worst_yearmonth'
    })
regionalResponseImWorst = regionalYoY.\
    iloc[regionalYoY.groupby('region')['im_vol'].idxmin()].\
    set_index('region').\
    rename(columns = {
        'yearmonth': 'im_worst_yearmonth'
    })

regionalResponseExLastDisrupt = regionalYoY.\
    query('ex_vol < 0').\
    reset_index(drop = True)
regionalResponseImLastDisrupt = regionalYoY.\
    query('im_vol < 0').\
    reset_index(drop = True)

regionalResponseExLastDisrupt = regionalResponseExLastDisrupt.\
    iloc[regionalResponseExLastDisrupt.groupby('region')['yearmonth'].idxmax()].\
    set_index('region').\
    rename(columns = {
        'yearmonth': 'ex_last_disrupt_yearmonth'
    })
regionalResponseImLastDisrupt = regionalResponseImLastDisrupt.\
    iloc[regionalResponseImLastDisrupt.groupby('region')['yearmonth'].idxmax()].\
    set_index('region').\
    rename(columns = {
        'yearmonth': 'im_last_disrupt_yearmonth'
    })

regionalResponseAfter = regionalYoY.\
    groupby(['region']).\
    aggregate({
        'ex_vol': 'max',
        'im_vol': 'max'
    }).\
    rename(columns = {
        'ex_vol': 'ex_response_after',
        'im_vol': 'im_response_after'
    })

regionalResilience = regionalRobust[['ex_robust', 'im_robust']].\
    join(regionalResponseExWorst[['ex_worst_yearmonth']]).\
    join(regionalResponseImWorst[['im_worst_yearmonth']]).\
    join(regionalResponseExLastDisrupt[['ex_last_disrupt_yearmonth']]).\
    join(regionalResponseImLastDisrupt[['im_last_disrupt_yearmonth']]).\
    join(regionalResponseAfter).\
    eval(
        'ex_response_recovery = (ex_last_disrupt_yearmonth + 1) - ex_worst_yearmonth' '\n'
        'im_response_recovery = (im_last_disrupt_yearmonth + 1) - im_worst_yearmonth'
    ).\
    eval(
        'ex_response_recovery = (24 - ex_response_recovery) / 24' '\n'
        'im_response_recovery = (24 - im_response_recovery) / 24'
    )
regionalResilience[['ex_robust', 'im_robust', 'ex_response_recovery', 'im_response_recovery', 'ex_response_after', 'im_response_after']].\
    to_excel('./Tables/RegionalResilience.xlsx')
```

## Country Scale

```python
countryScaleYoY = pandas.read_excel('./Tables/CountryScaleYoY.xlsx', sheet_name = 'CountryScaleYoY')
countryScaleYoY['yearmonth'] = (countryScaleYoY['year'] - 2020) * 12 + countryScaleYoY['month']

countryScaleRobust = countryScaleYoY.\
    groupby(['ISO3']).\
    aggregate({
        'ex_vol': 'min',
        'im_vol': 'min'
    }).\
    eval(
        'ex_robust = 1 + ex_vol' '\n'
        'im_robust = 1 + im_vol'
    )

countryScaleResponseExWorst = countryScaleYoY.\
    iloc[countryScaleYoY.groupby('ISO3')['ex_vol'].idxmin()].\
    set_index('ISO3').\
    rename(columns = {
        'yearmonth': 'ex_worst_yearmonth'
    })
countryScaleResponseImWorst = countryScaleYoY.\
    iloc[countryScaleYoY.groupby('ISO3')['im_vol'].idxmin()].\
    set_index('ISO3').\
    rename(columns = {
        'yearmonth': 'im_worst_yearmonth'
    })

countryScaleResponseExLastDisrupt = countryScaleYoY.\
    query('ex_vol < 0').\
    reset_index(drop = True)
countryScaleResponseImLastDisrupt = countryScaleYoY.\
    query('im_vol < 0').\
    reset_index(drop = True)

countryScaleResponseExLastDisrupt = countryScaleResponseExLastDisrupt.\
    iloc[countryScaleResponseExLastDisrupt.groupby('ISO3')['yearmonth'].idxmax()].\
    set_index('ISO3').\
    rename(columns = {
        'yearmonth': 'ex_last_disrupt_yearmonth'
    })
countryScaleResponseImLastDisrupt = countryScaleResponseImLastDisrupt.\
    iloc[countryScaleResponseImLastDisrupt.groupby('ISO3')['yearmonth'].idxmax()].\
    set_index('ISO3').\
    rename(columns = {
        'yearmonth': 'im_last_disrupt_yearmonth'
    })

countryScaleResponseAfter = countryScaleYoY.\
    groupby(['ISO3']).\
    aggregate({
        'ex_vol': 'max',
        'im_vol': 'max'
    }).\
    rename(columns = {
        'ex_vol': 'ex_response_after',
        'im_vol': 'im_response_after'
    })

countryScaleResilience = countryScaleRobust[['ex_robust', 'im_robust']].\
    join(countryScaleResponseExWorst[['ex_worst_yearmonth']]).\
    join(countryScaleResponseImWorst[['im_worst_yearmonth']]).\
    join(countryScaleResponseExLastDisrupt[['ex_last_disrupt_yearmonth']]).\
    join(countryScaleResponseImLastDisrupt[['im_last_disrupt_yearmonth']]).\
    join(countryScaleResponseAfter).\
    eval(
        'ex_response_recovery = (ex_last_disrupt_yearmonth + 1) - ex_worst_yearmonth' '\n'
        'im_response_recovery = (im_last_disrupt_yearmonth + 1) - im_worst_yearmonth'
    ).\
    eval(
        'ex_response_recovery = (24 - ex_response_recovery) / 24' '\n'
        'im_response_recovery = (24 - im_response_recovery) / 24'
    )
countryScaleResilience['ex_response_recovery'] = countryScaleResilience.apply(
    lambda row: 0 if row['ex_last_disrupt_yearmonth'] == 24 else row['ex_response_recovery'],
    axis = 1
)
countryScaleResilience['im_response_recovery'] = countryScaleResilience.apply(
    lambda row: 0 if row['im_last_disrupt_yearmonth'] == 24 else row['im_response_recovery'],
    axis = 1
)
countryScaleResilience['ex_response_recovery'] = countryScaleResilience.apply(
    lambda row: 1 if row['ex_robust'] >= 1 else row['ex_response_recovery'],
    axis = 1
)
countryScaleResilience['im_response_recovery'] = countryScaleResilience.apply(
    lambda row: 1 if row['im_robust'] >= 1 else row['im_response_recovery'],
    axis = 1
)
countryScaleResilience['ex_robust'] = countryScaleResilience.apply(
    lambda row: 1 if row['ex_robust'] >= 1 else row['ex_robust'],
    axis = 1
)
countryScaleResilience['im_robust'] = countryScaleResilience.apply(
    lambda row: 1 if row['im_robust'] >= 1 else row['im_robust'],
    axis = 1
)
countryScaleResilience[['ex_robust', 'im_robust', 'ex_response_recovery', 'im_response_recovery', 'ex_response_after', 'im_response_after']].\
    to_excel('./Tables/CountryScaleResilience.xlsx')
```
